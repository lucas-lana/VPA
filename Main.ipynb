```python
import os
import shutil
import zipfile
import kagglehub
```

```python
directory_path = '/content/drive/MyDrive/VPA_Project/Data'

# Check if the directory exists
if not os.path.exists(directory_path):
    # If it doesn't exist, create it. exist_ok=True prevents an error if the directory somehow gets created between the check and the creation attempt.
    os.makedirs(directory_path, exist_ok=True)
    print(f"Directory '{directory_path}' created successfully.")
else:
    print(f"Directory '{directory_path}' already exists.")
```

```python
# 1. Download the dataset (returns the path to the folder, NOT a zip file)
download_path = kagglehub.dataset_download("ryanluong1/valorant-champion-tour-2021-2023-data")
print(f"Dataset downloaded to cache at: {download_path}")

# 2. Define the destination directory in Google Drive
destination_directory = '/content/drive/MyDrive/VPA_Project/Data/VCT_Champions'

# 3. Copy the files from the cache to Google Drive
print(f"Copying files to {destination_directory}...")

try:
    # copytree copies the entire directory structure
    # dirs_exist_ok=True allows it to overwrite/merge if the folder already exists
    shutil.copytree(download_path, destination_directory, dirs_exist_ok=True)
    print("Success! Files copied.")
except Exception as e:
    print(f"An error occurred while copying: {e}")

# 4. Verify contents
print(f"\nContents of '{destination_directory}':")
if os.path.exists(destination_directory):
    for item in os.listdir(destination_directory):
        print(f"- {item}")
else:
    print("Directory does not exist.")
```
